```python
def add_features(row):
    df = dataset_global[
        (dataset_global['Genotype'] == row['Genotype']) 
        & (dataset_global['Position'] == row['Position']) 
        & (dataset_global['Codon'] == row['Codon'])]
    print("df", df)
    for item in Codon_Resistente_List:
        # Criando a linha de barreira gen√©tica
        print("Item:" ,item)
        gb = af.genetic_barrier(row[item],row['Codon'])
        column_name = "gb_" + item[-2:]
        row[column_name] = gb
        # Criando a linha de frequencia
        freq = df['Frequency'].values[0]
        codon_usage = df['HCV-G1'].values[0]
        column_name1 = "freq_" + item[-2:]
        row[column_name1] = freq
        column_name2 = "codon_usage_" + item[-2:]
        row[column_name2] = codon_usage
    return row

        
        
```

```python
modstep3 = step3.apply(lambda x: add_features(x), axis=1)
```
