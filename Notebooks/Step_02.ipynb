```python
genotype_1b = af.fasta_to_dict("../Data/" + "NS5A_1b_06.03.2018.aln")
genotype_1b_codons = {}
for sequence in genotype_1b.items():
    genotype_1b_codons[sequence[0]] = af.sequence_codons(sequence[1])
```

#### Contar as frequências

```python
df = pd.DataFrame(data=genotype_1b_codons)
df = df.transpose()
df_frequency = df.apply(pd.value_counts)
df_frequency = df_frequency.apply(lambda x: x/len(df))
```

```python
df_frequency
```

#### Tabela

```python
df_codons = pd.DataFrame(columns=["Genotype", "Position", "Codon", "Frequency"])
```

```python
for position in df_frequency.columns:
    for codon in df_frequency.index:
        new_row = [genotype, position, codon, df_frequency.at[codon,position]]
        new_series = pd.Series(new_row, index = df_codons.columns)
        df_codons = df_codons.append(new_series, ignore_index=True)
```
